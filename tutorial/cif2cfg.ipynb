# Notebook prepares a collection of cif files for structure repository


### Please modify a few settings as required

```python
# Description of the source of the cif files
source = 'Pearson Crystal Database'
# Descriptive one word abbreviation of source
sourceshort = 'Pearson'
# file type to look for
filetype = 'cif'
# top directory from where to start search for filetype
path = '/Users/drautz/worktodo/Pearson/downloads'
# path where to store database cfgs
targetpath = '/Users/drautz/worktodo/Pearson/converted'
```

```python
import ase
import glob
import os
import numpy as np
from amstools.resources.cfgio import create_structure,write_structure
```

```python
sourceshort = sourceshort.replace(" ", "")
filelist = glob.glob(path+'/**/*'+'.'+filetype, recursive=True)
if not os.path.exists(targetpath):
  os.mkdir(targetpath)
```

```python
for n in range(0,len(filelist)):
    file = filelist[n]
    basename = file.rsplit('/',1)[1].split('.',1)[0]
    print("Converting structure file:", file)
    at0 = ase.io.read(file)
    symbols = at0.get_chemical_symbols()
    formula = at0.get_chemical_formula()
    elements, counts = np.unique(symbols, return_counts=True)
    namedir = elements[0]
    for l in range(1,len(elements)):
      namedir = namedir+"-"+elements[l]
    savename = targetpath+'/'+namedir+'/'+sourceshort+'/'+basename+'.cfg'    
    if not os.path.exists(targetpath+'/'+namedir):
      os.mkdir(targetpath+'/'+namedir)
    if not os.path.exists(targetpath+'/'+namedir+'/'+sourceshort):
      os.mkdir(targetpath+'/'+namedir+'/'+sourceshort)
    at0.info["name"] = basename
    at0.info["alternative names"] = formula
    at0.info["source"] = source
    write_structure(savename, at0, save_exact=False,scaling_length=1.0)
    print("     Saved structure file:", savename)
```
