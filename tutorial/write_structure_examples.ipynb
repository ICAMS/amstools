# Save scaled structure

Any atoms object can be saved as an extended cfg with corresponding header and \[EleA, EleB, ...\] as chemical species

## Periodic strucs

The periodic cell is scaled down according to `alat` info given in table in structures repository https://git.noc.ruhr-uni-bochum.de/atomicclusterexpansion/data/structures

### Unary

```python
at0 = bulk("Mo", cubic=True)
write_structure("test_bcc.cfg", at0, save_exact=False)


with open("test_bcc.cfg","r") as fl:
    lines = fl.readlines()
lines
```

### Binary

```python
at0 = bulk("MoC",crystalstructure="rocksalt",a=3.15, cubic=True)
write_structure("test_rocksalt.cfg", at0, save_exact=False,scaling_length=1.0)


with open("test_rocksalt.cfg","r") as fl:
    lines = fl.readlines()
lines
```

## Non-periodic structures 

Atomic positions are not scaled and the structure is stored in a unit cell

```python
at0 = bulk("Mo", cubic=True)
at0.set_pbc(False)
write_structure("test_npbc.cfg", at0, save_exact=False)


with open("test_npbc.cfg","r") as fl:
    lines = fl.readlines()
lines
```

# Save structure exactly

Without scaling or changing the chemical species to EleA, EleB, etc

Useful for quickly writing application-specific cfg files for personal use (eg, from Active learning)

```python
at0 = bulk("MoC",crystalstructure="rocksalt",a=3.15, cubic=True)
# at0.set_pbc(False)
write_structure("test_exact.cfg", at0, save_exact=True)


with open("test_exact.cfg","r") as fl:
    lines = fl.readlines()
lines
```

# Dealing with metadata header

## Option 1: Have information in atoms.info

atoms.info is a dictionary where using the specific header tags, one can include info to be added to the cfg header

```python
at0 = bulk("Mo", cubic=True)

# data in `atoms.info` gets passed automatically to the header
at0.info["name"] = "bcc cubic structure"

write_structure("test_bcc.cfg", at0, save_exact=False)


with open("test_bcc.cfg","r") as fl:
    lines = fl.readlines()
    
lines
```

## Option 2: Pass a metadata_dict as argument (overwrites atoms.info!)

A separate dictionary can be passed as argument to the `write_structure` function. Its noted that the information passed here will overwrite atoms.info

`metadata_dict` should have the same tags (in smallercase) as in the header to be processed correctly

```python
at0 = bulk("Mo", cubic=True)

# data in `atoms.info` will be overwritten by manual_metadict
at0.info["name"] = "bcc cubic structure"

manual_metadict = {"author":"AMS",
                   "name":"BCC structure",
                   "periodic":"True",
                   "source":"ase_atoms_bulk",
                   "scaling_length":"1.0"}

write_structure("test_metadict.cfg", at0, save_exact=False,metadata_dict=manual_metadict)


with open("test_metadict.cfg","r") as fl:
    lines = fl.readlines()
    
lines
```
